### Import libraries

```python
import pandas as pd 
```

### Load data

```python
df=pd.read_csv('customers.csv')
```

```python
df1=pd.read_csv('orders.csv')
```

```python
df2=pd.read_csv('sales.csv')
```

```python
df3=pd.read_csv('products.csv')
```

### Merge the data

```python
df.columns
```

```python
first_Merge = pd.merge(left=df2, right=df3, left_on='product_id', right_on='product_ID', how='inner')
```

```python
second_Merge=pd.merge(left=df1, right=first_Merge,on='order_id', how='inner')
```

```python
df=pd.merge(left=df,right=second_Merge,on='customer_id',how='inner')
```

### what is the maximum number of days between the order and delivery


```python
(df['delivery_date'] - df['order_date']).dt.days.max()
```

### extract year , month and day name for order data


```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day_name'] = df['order_date'].dt.day_name() 
```

```python
df[['year','month','day_name']]
```

### extract year , month and day name for delivery data


```python
df['year'] = df['delivery_date'].dt.year
df['month'] = df['delivery_date'].dt.month
df['day_name'] = df['delivery_date'].dt.day_name() 
```

```python
df[['year','month','day_name']]
```

### what is the highest month that get the highest sales?
